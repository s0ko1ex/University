# 24. Собстсвенные значения и собственне векторы линейного пространства

Дать оперделение собственного вектора и собственного значения оператора, соответствующего собственному вектору. Дать значение спектра.

### Примеры

1. Для $ M_n $ и $ \mathcal D $ любой многочлен нулевой степени - собственный вектор с собственным значением равным 0
2. Оператор проектирования - вектора, принадлежащие линейный подпространствам $ L_1, L_2 $ - собственные вектора соотвествующий собственным значениям $ 0 $ и $ 1 $ соотвественно.

Любой собственный вектор порождает одномерное пространство собственных векторовЮ из которого исключён вектор $ \theta $.

### Теорема 85.1

Собственные векторы $ x_1, \dots, x_k $ оператора $ \mathcal A $ соотвествующие различным значениям $ \lambda_1, \dots, \lambda_k $ линейно независимы.

### Доказательство

Применим индукцию по $ k $. Для $ k = 1 $ это очевидно, т.к. по определению собственный вектор ненулевой.  
Пусть это векно для $ k - 1 $, докажем для $ k $. Допустим это не так, значит $ \exists \alpha_1, \dots, \alpha_k: \alpha_1 x_1 + \dots  + \alpha_k x_k = \theta $. Тогда, если применить к этой линейной комбинации оператор $ \mathcal A $, получаем $ \alpha_1 \lambda_1 x_1 + \dots + \alpha_n \lambda_n x_n = \theta $. Вычтем из этого изначальное равенство, домноженное на $ \lambda_n $. Получаем $ \alpha_1 (\lambda_1 - \lambda_n) x_1 + \dots + \alpha_{n - 1} (\lambda_{n - 1} - \lambda_n) x_{n - 1} = \theta $. Так как все собственные значения, получается, что линейная комбинация линейно независимых векторов даёт $ \theta $. Противоречие. Значит векторы $ x_1, \dots, x_k $ линейно независимы.
