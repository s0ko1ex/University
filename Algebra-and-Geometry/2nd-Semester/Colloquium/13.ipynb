# 13. Линейные операторы. Определение, основные свойства, примеры. Теорема о существовании и единственности оператора по заданным образам базисных векторов.

Пусть $ V, W $ - линейные пространства над общим полем $ P $. Отображение $ \mathcal{A}: V \to W $ называется линейным отоборажением пространства $ V $ а пространство $ W $, если $ \forall x, y \in V, \alpha \in P $
1. $ \mathcal{A}(x + y) = \mathcal A x + \mathcal A y $
2. $ \mathcal A (\alpha x) = \alpha \mathcal A x $

Это линейное отображение также называют линейным преобразованием пространства $ V $ в пространство $ W $ или линейным оператором, действующим из пространства $ V $ в пространство $ W $.

Если $ W = P $, то такое отображение называют линейным функционалом или линейной формой в протранстве $ V $.

### Примеры
1. $ \mathcal D p(t) = p'(t) $ - оператор дифференцирования в пространстве многочленов $ M_n $
2. $ M_\infty $ - пространство всех вещественных многочленов от одной переменной. Отображение $$ Sp(t) = \int\limits_0^t p(x)dx, $$ называется оператором интегрирования.
3. Операторы проектирования и отображения
4. Нулевой оператор
5. Тождественное отображение
6. Изоморфизм $ \varphi $ пространств $ V, W $

### Простейшие свойства

1. Линейный оператор переводит нулевой вектор в нулевой вектор:
   $ \mathcal A \theta_1 = \mathcal A 0x = 0 \mathcal A x = \theta_2, \space \theta_1 \in V, \theta_2 \in W $
2. Линейный оператор сохраняет линейные комбинации:
   $ \mathcal A (\sum_{i = 1}^k \alpha_i x_i) = \sum_{i = 1}^k \alpha_i \mathcal A x_i $
3. Линейный оператор сохраняет линейную зависимость, то есть переводит линейно зависимую систему векторов в линейно зависимую.

### Теорема 76.1

Пусть $ e_1, \dots, e_n $ - базис пространства $ V $, а $ g_1, \dots, g_n $ - произвольные векторы пространства $ W $. Тогда $ \exists ! \mathcal A \in \mathcal L(V, W) $, который переводит векторы $ e_1, \dots, e_n $ в векторы $ g_1, \dots, g_n $ соотвественно.

### Доказательство

Пусть $ x = \sum_{i = 1}^n x_i e_i \in V $. Тогда $ \mathcal Ax = x = \sum_{i = 1}^n x_i g_i $. Из единственности разложения по базису следует, что образ любого вектора обнозначно определён, при этом $ \mathcal e_i = g_i, \space i = \overline{1, n} $. Линейность данного оператора вытекает из линейности координат. Оператор единственен, так как если $ \mathcal B $ - любой другой оператор, переводящий векторы $ e_1, \dots, e_n $ в векторы $ g_1, \dots, g_n $, то $$ \mathcal B x = \mathcal B (\sum_{i = 1}^n x_i e_i) = \sum_{i = 1}^n x_i \mathcal B e_i = \sum_{i = 1}^n x_i g_i = \mathcal A x, \space \forall x \in V $$ Следовательно $ \mathcal{A = B} $

Profit.

Следствие. Операторы $ \mathcal {A, B \in L} (V, W) $ равны тогда и только тогда, когда они совпадают на векторах базиса $ V $.
